## Bayes' Theorem
#### - Naive Bayes Classifiers
#### - Spanners / Ranges - e.g. 2 machines producing the same parts (parts are marked by the machine that produced it)
#### - What is the probability, that a random data point has specific features?
#### - Bayes' Theorem mathmatical representation: P(A|B) = P(B|A) * P(A) / P(B)  
### - Example: Machine1 produces 30 parts/hr and Machine 20 parts 2/hr
##### - Out of all the produced parts: 1% are defective and out of these parts: 50% came from Machine1 and 50% from Machine2
##### - What is the probability of a part produced by Machine2 is defective?
##### - In total of all ranges produced there are 50 ranges produced per hour 
##### - P(Machine1) = 30/50 = 0.6 and P(Machine2) = 20/50 = 0.4
##### - P(Defect) = 1% - Probability of both machines producing a defect part
##### - P(Machine1 | Defect) = 50% - The probability of Machine1 producing a defect part | means given and here given the part produced by Machine1 is defect - Picked out from the defective pile - the subscription 
##### - P(Machine2 | Defect) = 50%
##### - From all the piles it would be 40% because Machine2 produces less parts than Machine2 and makes this machine more likely to produce a defect part since it only has a 0.4 probability but is still responsible for 50% of the defective parts produced
##### - P(Defect | Machine2) = ?
##### - P(Defect | Machine2) = P(Machine2 | Defect * P(Defect)) / P(Machine2) - Formula for calculating the example question
##### - P(Defect | Machine2) = 0.5 * 0.01 / 0.4 = 0.0125 or 1.25% propability of Machine2 producing a defective part
### Example: 1000 parts are produced in total

##### - 1% in total are defective = 10 parts
##### - Out the 1% defective parts total 50% of the parts was produced by Machine2

## Plan of Attack
### Given the features of X - what is the probability (Walks)?
### Step 1: Prior Probability
#### - Calculating #1 Prior Probability (Walks): Add new data point - what is the probability of this person (data point) walks to work?
##### - P(Walks) = Number of Walkers / Total Observations - P(Walks) = 10/30 for this example
### Step 2: Marginal Likelyhood
#### - Calculating #2 Marginal Likelyhood (Walks)
##### - Radius (circle) - Probability: P(X) = Number of Similar Observation / Total Observations - P(X) = 4/30
### Step 3: Likelihood 
#### - Calculating #3 Likelihood of random data point - Given | X Walks
##### - P(X|Walks) 
##### - P(X|Walks) = Number of Similar Observations Among those who Walk / Total number of Walkers - P(X|Walks) = 2/3
### Step 4: Posterior Probability
##### - Calculating #4 Posterior Probability: P(Walks|X) = 3/10 * 10/30 / 4/30 = 0.75 or 75% probability of the new data point feature to be classified as a Walker
### Probability (Drives)?
##### - Step 1: Posterior Probability for new data point feature to be classified as Drives: P(Drives|X) = 1/20 * 20/30 / 4/30 = 0.25 or 25%
##### - Step 2: Marginal Likelihood P(Walk|X ) v.s. P(Drives|X): 75% v.s. 25% - More likeli that a person  with features X Walks - P(Walks|X) > P(Drives|X)
##### - Step 3: Likelihood  of being similar to X Drives - P(X|Drives) = 1/20 
##### - Step 4: Posterior Probability - P(Drives|X) = 1/20 * 20/30 / 4/30 = 0.25 or 25%

# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Data Preprocessing

## Importing the Dataset

```python
dataset = pd.read_csv('Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the Train and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# The Naive Bayes Classification Model

## Training the Naive Bayes Model

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

### Predicting a new result

```python
classifier.predict(sc.transform([[30,87000]]))
```

### Predicting the Test set results

```python
# y_pred the matrix values of predictions
y_pred = classifier.predict(X_test)
np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)
```

# Visualizing The Naive Bayes Model

## The Training set results

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Naive Bayes (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

## The Test set results

```python
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Naive Bayes (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
